### DESAFIO:

*   Obter o faturamento da loja
*   Obter produtos por loja vendidos
*   Obter o ticket Medio das vendas



```python
# Importando o Pandas
import pandas as pd
```

```python
# Lendo nossa tabela
tabela = pd.read_excel('VendasAnalise3.xlsx')
display(tabela)
```

```python
# Fazendo com que a tabela fique sem nenhum dado indefinido
tabela = tabela.dropna(how='any', axis=0)
display(tabela)
```

```python
# Verificando se alguma tabela está com o tipo errado
tabela.info()
```

# Priemrio objetivo - Faturamento obitido

```python
faturamento_loja = tabela[['ID Loja', 'Valor Final']].groupby('ID Loja').sum().reset_index()
display(faturamento_loja)
print()
faturamento_total = tabela['Valor Final'].sum()
print('O faturamento total foi de: ', faturamento_total)
```

# Segundo Objetivo - Produtos vendidos por loja

```python
# Produtos vendidos por loja
Produtos_loja = tabela[['ID Loja', 'Produto', 'Quantidade']].groupby(['ID Loja', 'Produto']).sum().reset_index()
display(Produtos_loja)
```

```python
# Faturamento por loja
Produtos_fat = tabela[['ID Loja', 'Quantidade']].groupby('ID Loja').sum().reset_index()
display(Produtos_fat)
```

# Terceiro Objetivo - Ticket medio de cada loja

```python
ticket_medio = faturamento_loja['Valor Final']/Produtos_Loja['Quantidade']
display(ticket_medio)
```

```python
tabela_ticket = pd.DataFrame({'Lojas': Produtos_Loja['ID Loja'], 'Ticket Medio' : ticket_medio})
display(tabela_ticket)
```
